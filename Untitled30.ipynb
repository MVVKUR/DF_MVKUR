```python
from transformers import pipeline

checkpoint = "openai/clip-vit-base-patch32"
detector = pipeline(model=checkpoint, task="zero-shot-image-classification")
```

```python
from PIL import Image

# Path to the local PNG image file
png_path = "/content/drive/MyDrive/iden/66bac7b85cd3ca8621e3ce5235a86560-SD_IMAGE IS NOT KTP-REC_----TC_--SRC_--MC_--RR_--modified.png"

# Open the image
image = Image.open(png_path)

# Display the image
image
```

```python
predictions = detector(image, candidate_labels=["blured", "croped", "Foto bukan ktp", "Foto adalah KTP"])
image
predictions
```
